```python
import matplotlib.pyplot as plt

# Datos experimento
muestra1 = [35, 42, 58, 65, 48, 52, 60, 38, 62, 45]
muestra2 = [48, 52, 49, 51, 50, 53, 47, 51, 52, 50]
muestra3 = [49, 51, 50, 52, 50, 49, 51, 50, 51, 50]

# Crear la figura y los ejes
plt.figure(figsize=(8, 6))

# Graficar los datos
plt.scatter(muestra1, [1]*len(muestra1), marker='*', label='Muestra 1')
plt.scatter(muestra2, [2]*len(muestra2), marker='o', label='Muestra 2')
plt.scatter(muestra3, [3]*len(muestra3), marker='.', label='Muestra 3')

# Personalizar la gráfica
plt.xlabel('Valores')
plt.ylabel('Muestras')
plt.title('Distribución de muestras')
plt.yticks([1, 2, 3], ['Muestra 1', 'Muestra 2', 'Muestra 3'])
plt.xlim(30, 70)
plt.legend()

# Mostrar la gráfica
plt.show()
```
