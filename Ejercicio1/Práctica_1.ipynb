```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# Datos de temperatura
temperaturas = [84, 49, 61, 40, 83, 67, 45, 66, 70, 69,
                80, 58, 68, 60, 67, 72, 73, 70, 57, 63,
                70, 78, 52, 67, 53, 67, 75, 61, 70, 81,
                76, 79, 75, 76, 58, 31]

# Cálculo de medidas de tendencia central
media = np.mean(temperaturas)
mediana = np.median(temperaturas)

# Cálculo de la moda usando Counter
contador = Counter(temperaturas)
moda = contador.most_common(1)[0][0]  # Obtener el valor más común

print(f'Media: {media:.2f}')
print(f'Mediana: {mediana:.2f}')
print(f'Moda: {moda:.2f}')

# Cálculo del rango y desviación estándar
rango = np.max(temperaturas) - np.min(temperaturas)
desviacion_estandar = np.std(temperaturas)

print(f'Rango: {rango}')
print(f'Desviación Estándar: {desviacion_estandar:.2f}')

# Crear un histograma
plt.hist(temperaturas, bins=10, color='skyblue', edgecolor='black')
plt.title('Distribución de Temperaturas de los Sellos Anulares')
plt.xlabel('Temperatura (°F)')
plt.ylabel('Frecuencia')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Identificación de la temperatura más baja y la más alta
temp_min = np.min(temperaturas)
temp_max = np.max(temperaturas)

# Calcular la diferencia entre la temperatura más baja y la siguiente más baja
temperaturas.sort()  # Ordenar para encontrar la siguiente más baja
diferencia = temperaturas[1] - temp_min

print(f'Temperatura más baja: {temp_min}')
print(f'Temperatura más alta: {temp_max}')
print(f'Diferencia entre la temperatura más baja y la siguiente más baja: {diferencia}')

# Análisis de temperaturas críticas
umbral = 53
porcentaje_bajo_umbral = np.sum(np.array(temperaturas) < umbral) / len(temperaturas) * 100

print(f'Porcentaje de lanzamientos por debajo de {umbral}°F: {porcentaje_bajo_umbral:.2f}%')


```
